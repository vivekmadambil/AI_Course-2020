**Rabbit Leap Problem **

```
START = "LL-RR"
GOAL  = "RR-LL"

# Swap function
def swap_positions(s,i,j):
  assert i < j
  return s[:i] + s[j] + s[i+1:j] + s[i] + s[j+1:]

# Function to get valid successor nodes
def getchild(p):
  i = p.index('-')
  cs = []
  c1=p
  c2=p
  c3=p
  c4=p
  # Four operations
  if i-2 >= 0 and p[i-2]=='L':
    cs.append(swap_positions(c1,i-2,i))
  if i-1 >= 0 and p[i-1]=='L':
    cs.append(swap_positions(c2,i-1,i))
  if i+1 < len(p) and p[i+1]=='R':
    cs.append(swap_positions(c3,i,i+1))
  if i+2 < len(p) and p[i+2]=='R':
    cs.append(swap_positions(c4,i,i+2))
  return cs

# DFS Algorithm   
def dfs_search(s):
  SL = list([(s)])
  NSL = list([(s)])
  DE = list()
  CS =s
  state = 1
  nochild = 0
  chid_count = 0
  while NSL!=0:
    print("\n***")
    state=state+1
    # Check GOAL state
    if CS == GOAL:
      print("SL:",SL) # contains path from goal state to start state
      print("No of states explored:",state)
      print("Level of Goal State:", len(SL)-1)
      return
    print("Current Node:", CS)
    child_list = list([""]) 
    child_list = getchild(CS) # Function to get valid successor nodes
    print("Children of", CS, "is", child_list)
    chid_count=len(child_list)
    print("DE:",DE,"\nSL:", SL,"\nNSL:", NSL)
    #check if child present in DE, SL and NSL
    for child in child_list:
      for t1 in DE:
        if child==t1:
          chid_count=chid_count-1
          break
      for t2 in SL:
        if child==t2:
          chid_count=chid_count-1
          break
      for t3 in NSL:
        if child==t3:
          chid_count=chid_count-1
          break
    print("chid_count(after check):",chid_count)
    # If No children (excluding DE,SL and NSL)
    if chid_count==0:
      print("Reached Dead End") 
      #Backtracking
      while SL!=0 and CS == SL[0]:
        DE.insert(0,CS)
        s=SL[0]
        SL.remove(s)
        s=NSL[0]
        NSL.remove(s)
        CS = NSL[0]
      SL.insert(0,CS)
      print("After Dead End Process\nDE:",DE,"\nSL:", SL,"\nNSL:", NSL)    
    else:
      # Remove children already in DE, SL and  NSL
      for child in child_list:
        for t4 in DE:
          if child==t4:
            child_list.remove(child)
            break
        for t5 in SL:
          if child==t5:
            child_list.remove(child)
            break
        for t6 in NSL:
          if child==t6:
            child_list.remove(child)
            break
      # New child node identified
      print("child_list:",child_list)
      for child in child_list:
        NSL.insert(0,child)
      CS=NSL[0]  
      SL.insert(0,CS)

# Main program
print("Rabbit Leap Problem using DFS\n Start State:", START,"Goal State:", GOAL)
dfs_search(START)
```
