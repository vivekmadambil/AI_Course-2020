```
START = 'ABC-XYZ'
GOAL  = 'XYZ-ABC'
RFROGS = 'ABC'
LFROGS = 'XYZ'

def swap_positions(s,i,j):
    assert i < j
    return s[:i] + s[j] + s[i+1:j] + s[i] + s[j+1:]

def apply_move(s,f):
    i = s.index(f)
    if f in RFROGS: # go right
        if i+1 < len(s) and s[i+1] == '-':
            return swap_positions(s,i,i+1)
        if i+2 < len(s) and s[i+1] != '-' and s[i+2] == '-':
            return swap_positions(s,i,i+2)
    if f in LFROGS: # go left
        if i-1 >= 0 and s[i-1] == '-':
            return swap_positions(s,i-1,i)
        if i-2 >= 0 and s[i-1] != '-' and s[i-2] == '-':
            return swap_positions(s,i-2,i)

def show_state(s):
    print(s)

def next_moves(s):
    i = s.index('-')
    ms = []
    # in from the left
    if i-1 >= 0 and s[i-1] in RFROGS:
        ms.append(swap_positions(s,i-1,i))
    if i-2 >= 0 and s[i-2] in RFROGS:
        ms.append(swap_positions(s,i-2,i))
    # in from the right
    if i+1 < len(s) and s[i+1] in LFROGS:
        ms.append(swap_positions(s,i,i+1))
    if i+2 < len(s) and s[i+2] in LFROGS:
        ms.append(swap_positions(s,i,i+2))
    return ms

def search(s):
    work = set([("",s)])
    states = 0
    while work:
        path, cur = work.pop()
        states += 1
        show_state(cur)
        for k in next_moves(cur):
            succ = apply_move(cur, k)
            work.add((path+k, succ))
            print("  ", path+k, " --> ",)
            show_state(succ)
            if succ == GOAL:
                print("           !!! GOAL in", len(path)+1, " moves.")
                return
    print("Visited", states, "states.")

search(START)
```

```
START = "LLL-RRR"
GOAL  = "RRR-LLL"


def swap_positions(s,i,j):
  assert i < j
  return s[:i] + s[j] + s[i+1:j] + s[i] + s[j+1:]


def getchild(p):
  i = p.index('-')
  cs = []
  c1=p
  c2=p
  c3=p
  c4=p
  if i-2 >= 0 and p[i-2]=='L':
    cs.append(swap_positions(c1,i-2,i))
  if i-1 >= 0 and p[i-1]=='L':
    cs.append(swap_positions(c2,i-1,i))
  if i+1 < len(p) and p[i+1]=='R':
    cs.append(swap_positions(c3,i,i+1))
  if i+2 < len(p) and p[i+2]=='R':
    cs.append(swap_positions(c4,i,i+2))
  return cs
   
def dfs_search(s):
  SL = list([(s)])
  NSL = list([(s)])
  DE = list()
  CS =s
  
  #print(SL, NSL, DE, CS)
  state = 0
  nochild = 0
  ccount = 0
  while NSL!=0:
    print("\nLoop Start")
    state=state+1
    if CS == GOAL:
      print("SL:",SL)
      print("No of states explored:",state)
      return
    print("Current Node:", CS)
    GC = list([""]) 
    GC = getchild(CS)
    print("Children of", CS, "is", GC)
    ccount=len(GC)
    print("DE:",DE,"\nSL:", SL,"\nNSL:", NSL)
    for child in GC:
      for t1 in DE:
        if child==t1:
          ccount=ccount-1
          break
      for t2 in SL:
        if child==t2:
          ccount=ccount-1
          break
      for t3 in NSL:
        if child==t3:
          ccount=ccount-1
          break
    print("Ccount(after check):",ccount)
    if ccount==0:
      print("Reached Dead End") 
      while SL!=0 and CS == SL[0]:
        DE.insert(0,CS)
        s=SL[0]
        SL.remove(s)
        s=NSL[0]
        NSL.remove(s)
        CS = NSL[0]
      SL.insert(0,CS)
      print("After Dead End Process\nDE:",DE,"\nSL:", SL,"\nNSL:", NSL)    
    else:
      for child in GC:
        for t4 in DE:
          if child==t4:
            GC.remove(child)
            break
        for t5 in SL:
          if child==t5:
            GC.remove(child)
            break
        for t6 in NSL:
          if child==t6:
            GC.remove(child)
            break
      print("GC:",GC)
      for child in GC:
        NSL.insert(0,child)
      CS=NSL[0]  
      SL.insert(0,CS)
```

```
# BFS Algorithm   
def bfs_search(s):
  SL = list([(s)])
  NSL = list([(s)])
  DE = list()
  CS =s
  
  #print(SL, NSL, DE, CS)
  state = 1
  nochild = 0
  chid_count = 0
  while NSL!=0:
    print("\nLoop Start")
    state=state+1
    # Check GOAL state
    if CS == GOAL:
      print("SL:",SL)
      print("No of states explored:",state)
      print("Level of Goal State:", len(SL)-1)
      return
    print("Current Node:", CS)
    GC = list([""]) 
    GC = getchild(CS) # Function to get valid successor nodes
    print("Children of", CS, "is", GC)
    chid_count=len(GC)
    print("DE:",DE,"\nSL:", SL,"\nNSL:", NSL)
    #check if child present in DE, SL and NSL
    for child in GC:
      for t1 in DE:
        if child==t1:
          chid_count=chid_count-1
          break
      for t2 in SL:
        if child==t2:
          chid_count=chid_count-1
          break
      for t3 in NSL:
        if child==t3:
          chid_count=chid_count-1
          break
    print("chid_count(after check):",chid_count)
    # If No children (excluding DE,SL and NSL)
    if chid_count==0:
      print("Reached Dead End") 
      #Backtracking
      while SL!=0 and CS == SL[0]:
        DE.insert(0,CS)
        s=SL[0]
        SL.remove(s)
        s=NSL[0]
        NSL.remove(s)
        CS = NSL[0]
      SL.insert(0,CS)
      print("After Dead End Process\nDE:",DE,"\nSL:", SL,"\nNSL:", NSL)    
    else:
      # Remove children already in DE, SL and  NSL
      for child in GC:
        for t4 in DE:
          if child==t4:
            GC.remove(child)
            break
        for t5 in SL:
          if child==t5:
            GC.remove(child)
            break
        for t6 in NSL:
          if child==t6:
            GC.remove(child)
            break
      # New child node identified
      print("GC:",GC)
      for child in GC:
        NSL.append(child)
      NSL.remove(CS)
      CS=NSL[0]  
      SL.insert(0,CS)

# Main program
print("Start State:", START,"Goal State:", GOAL)
bfs_search(START)
```
